### 1. What is name of car that has the highest horsepower?  

```python
if 'level_0' in car.columns:
    car.drop(columns=['level_0'], inplace=True)
car.reset_index(inplace=True)
car_name = car.loc[car['horsepower'].idxmax(), 'name']
print("Car with highest horsepower:", car_name)
```

### 2. How many cars have mpg ≥ 35?

```python
count = (car['mpg'] >= 35).sum()
print("Number of cars with MPG >= 35:", count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
most_common_origin = car[(car['horsepower'] > 100) & (car['weight'] < 3000)]['origin'].mode()[0]
print("Most common origin:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_acc_japan = car[car['origin'] == 'japan']['acceleration'].mean()
print(f"Mean acceleration of cars from Japan: {mean_acc_japan:.2f}")
```

### 5. Which year had the highest average mpg?

```python

highest_avg_mpg_year = car.groupby('model_year')['mpg'].mean().idxmax()
print(f"Year with highest average mpg: {highest_avg_mpg_year}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
mpg_median = car['mpg'].median()
above_median_mpg_cars = car[car['mpg'] > mpg_median].copy()
above_median_mpg_cars['hp_to_weight'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']
max_ratio = above_median_mpg_cars['hp_to_weight'].max()
best_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight'] == max_ratio]
print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg_by_year_origin = car.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.ylabel('Average MPG')
plt.xlabel('Year')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
df = car.copy()
counts = df.groupby('name')['model_year'].nunique()
multi_year_names = counts[counts > 1].index
filtered = df[df['name'].isin(multi_year_names)]
stats = filtered.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
)
result = stats[(stats['mpg_std'] < 1.0)].sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])
print(result)
```
